```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
length = 10
k = 0.89
total_Time = 10

# BC
temp_L = 100
temp_R = 200
```

```python
dx = 0.1
x_values = np.linspace(0, length, int(length/dx))
print(x_values)
x_values.dtype
```

```python
dt = 0.0001
t_values = np.linspace(0, length, int(total_Time/dt))
print(t_values)
t_values.dtype
```

```python
# IC

u = np.zeros((len(t_values), len(x_values)))
print(u)
u.shape
```

```python
u[:,0] = temp_L
u[:, -1] = temp_R
```

```python
# FDM, forward time, central space

for t in range(1, len(t_values) -1):
    for x in range(1, len(x_values) -1):
        u[t+1,x] = u[t, x] + (k*(dt/dx**2)) * (u[t, x+1] - 2*u[t, x] + u[t, x-1])
```

```python
len(u)
```

```python
for i in range(0, len(u), int(len(u)/10)):
    plt.plot(x_values, u[i])

plt.show()
```
